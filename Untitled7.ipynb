```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency,ttest_ind

sns.set(style="whitegrid")
```

```python
v = pd.read_csv("/content/Titanic-Dataset.csv")
v.head()
```

```python
v.info()
v.describe(include='all')
```

```python
v.isnull().sum()
```

```python
sns.countplot(x="Survived",data=v)
plt.title("Survival Count")
plt.show()

sns.histplot(v['Age'].dropna(),kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x="Sex", y="Survived", data=v)
plt.title("Survival Rate by Gender")
plt.show()

sns.barplot(x="Pclass", y="Survived", data=v)
plt.title("Survival Rate by Class")
plt.show()

sns.boxplot(x="Survived", y="Age", data=v)
plt.title("Age Distribution by Survival")
plt.show()
```

```python
male_survival=v[v['Sex'] == 'male']['Survived']
female_survival=v[v['Sex'] == 'male']['Survived']
t_stat,p_val = ttest_ind(female_survival,male_survival)
print("T-statistic",t_stat, "| P-value",p_val)
if p_val<0.05:
  print("Statistically significant difference between male and female survival")
else:
  print("No significant difference")
```

```python
contingency = pd.crosstab(v['Pclass'], v['Survived'])

chi2, p, dof, expected = chi2_contingency(contingency)

print("Chi-square:", chi2, "| P-value:", p)

if p < 0.05:
    print("Passenger class has a significant effect on survival.")
else:
    print("No significant relationship.")
```

```python
v["AgeGroup"] = pd.cut(
    v["Age"],
    bins=[0, 12, 18, 65, 80],
    labels=["Child", "Teen", "Adult", "Senior"]
)

sns.barplot(x="AgeGroup", y="Survived", data=v)
plt.title("Survival Rate by Age Group")
plt.show()
```
